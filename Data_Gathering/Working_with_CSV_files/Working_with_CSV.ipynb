## Opening a CSV file from a URL or from a Server

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv" 
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"} #need to set out headers to send a request to CDN
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

### Converting Columns Data to our Desired need


```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
print(rename("Royal Challengers Bangalore"))
```

```python
pd.read_csv('./Downloads/IPL Matches 2008-2020.csv',converters={'team1':rename}) #So now Royal Challengers Bangalore gets converetd to RCB in Team1 Column
```

### Na_Values Parameters => Here we define what our custom NAN values can look like for ex - "-", "NaN", " "(space), etc and those gets converetd to NAN for our Parsed Data

```python
pd.read_csv('./Downloads/aug_train.csv',na_values=['Male',]) # here Value "Male" gets converted to "NAN" value
```

## Loading Huge Dataset in Chunks

```python
dataFrames = pd.read_csv('./Downloads/aug_train.csv', chunksize=4500)
```

```python
for chunks in dataFrames:
    print(chunks.shape) #now thourgh looping we access our chunks and perform our operation accordingly so that we dont have memory issues
```
