## 4. OneHotEncoding with Top Categories - for using Frequent Values and Other categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 80 #any brand below threshold value will get added to the "others-uncommon" category
```

```python
repl = counts[counts <= threshold].index
repl
```

```python
pd.get_dummies(df['brand'].replace(repl, 'others-uncommon')).sample(5)
```
