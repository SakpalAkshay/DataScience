```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
'''
Attribute Information:

Age of patient at time of operation (numerical)
Patient's year of operation (year - 1900, numerical)
Number of positive axillary nodes detected (numerical)
Survival status (class attribute)
1 = the patient survived 5 years or longer
2 = the patient died within 5 year
'''

'''What we need to do is this is a classification problem where we use variables or features such as Age, Patient Year of Operation
and Number of node to detect to classify if patient survied 2 or patient died 1'''
```

```python
# reading the file 
haberman = pd.read_csv('haberman.csv')
haberman.shape
```

```python
#lets see how the data looks like
haberman.head(10)
```

```python
#renaming our column names
#haberman.columns # => Index(['30', '64', '1', '1.1'], dtype='object')
new_columns = ['age','year','node','status']
haberman.columns = new_columns
```

```python
haberman.value_counts('status') #there is approx 70% 30% imbalance of data as number of patient who lived, is like 3 times more than who died
```

```python
#lets plot a 2-D scatter plot against nodes present and status
haberman.plot( kind ="scatter", x = "status", y = "node")
plt.show()
```

```python
#lets make it more differenciable 
sns.set_style('whitegrid')
sns.FacetGrid(haberman, hue='status') \
    .map(plt.scatter, 'node','status') \
    .add_legend();
plt.show(); #doesnt make it clear anything
```

```python
#Pair Plot, lets utilize them
sns.set_style("whitegrid");
sns.pairplot(haberman,hue='status',height= 4);
plt.show() #from here we can say that age and node have one siginificance thing that any person below 40 and node less than 20 has very high 
#% chance of survival
```

```python
sns.set_style("whitegrid");
#size -> size of the graph
sns.FacetGrid(haberman, hue='status',height=5) \
    .map(plt.scatter, 'node','age') \
    .add_legend();
plt.show();
```

## Univariate Analysis

```python
#Univariate Analysis through Histogram or displot
#Univariate using Age

sns.FacetGrid(haberman, hue="status", height = 5).map(sns.distplot, 'age').add_legend()
plt.show() #show to much overlapping not good for analysis
```

```python
#Univariate using year
sns.FacetGrid(haberman, hue="status", height = 5).map(sns.distplot, 'year').add_legend()
plt.show() #shows to much overlapping not good for analysis
```

```python
#Univariate using nodes
sns.FacetGrid(haberman, hue="status", height = 5).map(sns.distplot, 'node').add_legend()
plt.show() #shows how people with less number of nodes are the ones having more survival rate
```

```python
#Plotting CDF or Cumulative distributed Function
survived = haberman.loc[haberman["status"] == 1];
not_survived = haberman.loc[haberman["status"] == 2];
```

```python
counts, bin_edges = np.histogram(survived['node'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)



plt.show()
```

```python
counts, bin_edges = np.histogram(not_survived['node'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)



plt.show() #Around 80% people with node less 5 than survival rate
```
