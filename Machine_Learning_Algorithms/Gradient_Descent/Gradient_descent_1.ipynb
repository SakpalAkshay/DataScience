```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples=5, n_features=1, noise=80, random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
```

```python
#lets apply gradient descent assuming our slope M is constant = -3.175,  we also assume our intercept is 0
yPred = (-3.175 * X + 0).reshape(5) # because we had 5 samples here
```

```python

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,yPred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
#lets use b = 100
yPred = (-3.175 * X + 100).reshape(5)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,yPred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
#calculating b new from b old  - learning rate * slope
m = -3.175
b = 100

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((m * X) + b).reshape(5)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,yPred,color='#A3E4D7',label='b = 100')
plt.legend()
plt.show()
```

```python
#now use epochs and see how well it looks like with learning rate

b = -100
m = -3.175
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
